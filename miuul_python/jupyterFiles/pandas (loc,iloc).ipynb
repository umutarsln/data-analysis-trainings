```python
import pandas as pd
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
df = sns.load_dataset("titanic")
```

```python
df.iloc[0:3]
```

```python
df.loc[0:3]
```

```python
df.loc[0:3,"fare"]
```

```python
(df["age"]>40).head()
```

```python
df[df["age"]>40].head()
```

```python
df[df["age"]>40][["age","fare","class"]].head()
```

```python
df.loc[df["age"] > 40, ["age","fare","class"]].head()
```

```python
df.loc[(df["age"] > 40) & (df["sex"] == "male"), ["age","fare","class"]].head()
```

```python
df["age"].mean()
```

```python
# Gives pandas series
df.groupby("sex")["age"].mean()
```

```python
# Gives dataframe
df.groupby("sex").agg({"age":["mean","sum"]})
```

```python
df.groupby("sex").agg({"age":["mean","sum"], 
                      "survived":"mean"})
```

```python
df.groupby(["sex","embark_town"]).agg({"age":["mean"], 
                      "survived":"mean"})
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age":["mean"], 
                      "survived":"mean"})
```

```python
# Default calculates mean
df.pivot_table("survived","sex","embark_town")
```

```python
df.pivot_table("survived","sex","embark_town", aggfunc="sum")
```

```python
df.pivot_table("survived",["sex"],["embark_town","class"])
```

```python
df["new_age"] = pd.cut(df["age"],[0,20,40,60,90])
```

```python
df.pivot_table("survived",["sex","new_age"],["embark_town","class"])
```
